```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
train_df=pd.read_csv("/content/drive/MyDrive/fraudTrain.csv")
train_df.head()
```

```python
test_df=pd.read_csv("/content/drive/MyDrive/fraudTest.csv")
test_df.head()
```

```python
combined_df=pd.concat([train_df, test_df], ignore_index=True)
data=combined_df
```

```python
# Display the column names in your dataset
print(data.columns)
```

```python
data['trans_date_trans_time']= pd.to_datetime(data['trans_date_trans_time'])
data['hour'] = data['trans_date_trans_time'].dt.hour
data['day_of_week'] = data['trans_date_trans_time'].dt.dayofweek
data['month']= data['trans_date_trans_time'].dt.month
data = data.drop(['Unnamed: 0','trans_date_trans_time','merchant', 'first', 'last', 'street','city', 'state', 'zip','job','dob','trans_num'], axis=1)
data= pd.get_dummies(data, columns=['category'], drop_first=True)

data['gender'] = data['gender'].map({'M': 0, 'F': 1})
```

```python
X = data.drop(['is_fraud'], axis=1)
Y = data ['is_fraud']
scaler = StandardScaler()
numerical_columns = ['amt', 'lat', 'long', 'city_pop','unix_time', 'merch_lat', 'merch_long']
X[numerical_columns]= scaler.fit_transform(X[numerical_columns])
```

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
classifier= DecisionTreeClassifier(max_depth=4)

classifier.fit(X_train, Y_train)
prediction = classifier.predict(X_test)
print("\n Predicted value:\n",prediction)
```

```python
from sklearn import metrics
DecisionTree= metrics.accuracy_score(Y_test,prediction)*100
print("Accuracy score:",DecisionTree)
```
